### **#12 - Plot predictions**

```python
# fetch predict set
valid = np.load('Input/valid_patch128.npy')
# fetch predictions
predictions = np.load('Output/Model/predicts.npy')
predictions[predictions>=0.35] = 1
predictions[predictions<0.35] = 0
# plot
fig, axes = plt.subplots(10,5, figsize=(128,128))
for i,ax in enumerate(axes.flat):
    ax.imshow(predictions[i][:,:,1])
    ax.imshow(valid[i], alpha=0.5, cmap='gray')
```
